### 🔍 Checking for Duplicate Rows

When working with real-world datasets, duplicate rows are common and need to be identified before further analysis. Pandas provides the `.duplicated()` function, which helps us detect them:

1. **`df.duplicated().sum()`**  
   - This counts the number of **duplicate rows excluding the first occurrence**.  
   - Example: if a row appears 3 times, the first is considered original, and the next 2 are counted as duplicates.  

2. **`df.duplicated(keep=False).sum()`**  
   - This counts **all instances of duplicate rows, including the first occurrence**.  
   - Example: if a row appears 3 times, all 3 will be counted.  

👉 Using both gives us a complete picture of how many duplicates exist in the dataset.


````python
# Count duplicate rows (excluding first occurrence)
df.duplicated().sum()

# Count *all* duplicates (including the first one)
df.duplicated(keep=False).sum()
````

### 📝 Duplicate Report Function

To streamline duplicate detection, we define a **custom function** `duplicate_report(df)` that generates a structured report.  

The report includes:  
1. **Full row duplicates** → Total number of rows that are repeated.  
2. **Column-wise duplicates** → For each column, counts how many repeated values exist.  
3. **Top duplicate values per column** → Shows the most frequent values (top 3) in each column that have duplicates.  
4. **Duplicate rows preview** → Displays the first few duplicate rows for inspection.  

This gives us both a **row-level** and **column-level** understanding of duplication in the dataset.  


````python
def duplicate_report(df):
    print("🔎 DUPLICATE ANALYSIS REPORT")
    print("="*40)

    # 1. Full row duplicates
    full_row_dups = df.duplicated().sum()
    print(f"\n➡ Full row duplicates: {full_row_dups}")

    # 2. Column-wise duplicates
    print("\n➡ Column-wise duplicates:")
    any_col_dup = False
    for col in df.columns:
        dups = df[col].duplicated().sum()
        if dups > 0:
            any_col_dup = True
            print(f"   - {col}: {dups}")
    if not any_col_dup:
        print("   (No column-wise duplicates found)")

    # 3. Top duplicate values per column
    print("\n➡ Top duplicate values per column:")
    any_top = False
    for col in df.columns:
        counts = df[col].value_counts()
        if counts.max() > 1:  # only show if duplicates exist
            any_top = True
            print(f"\n   {col}:")
            print(counts.head(3))  # top 3
    if not any_top:
        print("   (No repeating values found)")

    # 4. Duplicate rows (if any)
    duplicate_rows = df[df.duplicated(keep=False)]
    print("\n➡ Duplicate rows preview:")
    if duplicate_rows.empty:
        print("   (No duplicate rows found)")
    else:
        print(duplicate_rows.head())
````

### 🔑 Identifying Key Columns

In relational databases or clean datasets, there is usually at least one **unique identifier** (like `ID`, `Accident_ID`, etc.) that can uniquely identify each row.  
Sometimes, such a key is not a single column but a **combination of multiple columns**.  

The function `find_key_columns(df, max_combo=3)` helps us automatically detect such key(s):  

1. **Single columns** → Checks if any column has all unique values (no duplicates).  
2. **Combinations** → If no single column works, it tests combinations of 2 or 3 columns (configurable with `max_combo`).  
3. **Result** → Returns a list of candidate key columns (or combinations) and prints them clearly.  

This ensures we know if our dataset has a **primary key** or whether duplicates are unavoidable.

````python
from itertools import combinations

def find_key_columns(df, max_combo=3):
    """
    Identify columns that can serve as unique keys in the DataFrame.
    1. Checks single columns.
    2. If none are unique, checks combinations of 2 or 3 columns.
    
    Parameters:
        df : pandas.DataFrame
        max_combo : int (default=3)
            Maximum size of column combinations to test.
    
    Returns:
        key_candidates : list
            List of columns or column combinations that can act as unique keys.
    """
    total_rows = len(df)
    key_candidates = []
    
    # 1. Check single columns
    for col in df.columns:
        if df[col].nunique(dropna=False) == total_rows:
            key_candidates.append([col])
    
    # 2. If no single key found, check combinations up to max_combo
    if not key_candidates:
        for r in range(2, max_combo + 1):
            for combo in combinations(df.columns, r):
                if df[list(combo)].drop_duplicates().shape[0] == total_rows:
                    key_candidates.append(list(combo))
    
    # Print results
    if key_candidates:
        print("✅ Key column(s) or combinations found:")
        for combo in key_candidates:
            print("   -", combo)
    else:
        print("❌ No unique column or combination up to", max_combo, "columns.")
    
    return key_candidates
````

````python
def unique_report(df):
    print("🔎 UNIQUE VALUE REPORT")
    print("="*40)
    total_rows = len(df)

    for col in df.columns:
        n_unique = df[col].nunique(dropna=False)
        is_key = "✅ Key Candidate" if n_unique == total_rows else ""
        print(f"{col:<25} → {n_unique} unique values {is_key}")
````
